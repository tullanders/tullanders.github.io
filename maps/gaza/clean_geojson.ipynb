```python
import json
with open('gaza_2024-09-06.geojson') as f:
    geojson_data = json.load(f)


oldFeatures = geojson_data["features"]
newFeatures = []


for feature in oldFeatures:
    newFeature = {}
    properties = feature["properties"]

    sensors =  []
    sensorDates = []
    confidences = []
    damageStatus = []
    mainDamageSiteClasses = []

    sensors.append(properties.get("SensorID"))
    sensors.append(properties.get("SensorID_2"))
    sensors.append(properties.get("SensorID_3"))
    sensors.append(properties.get("SensorID_4"))
    sensors.append(properties.get("SensorID_5"))
    sensors.append(properties.get("SensorID_6"))
    sensors.append(properties.get("SensorID_7"))
    sensors.append(properties.get("SensorID_8"))
    sensors.append(properties.get("SensorID_9"))


    # sensorDates
    sensorDates.append(properties.get("SensorDate"))
    sensorDates.append(properties.get("SensorDate_2"))
    sensorDates.append(properties.get("SensorDate_3"))
    sensorDates.append(properties.get("SensorDate_4"))
    sensorDates.append(properties.get("SensorDate_5"))
    sensorDates.append(properties.get("SensorDate_6"))
    sensorDates.append(properties.get("SensorDate_7"))
    sensorDates.append(properties.get("SensorDate_8"))
    sensorDates.append(properties.get("SensorDate_9"))

    # Confidences
    confidences.append(properties.get("ConfidenceID"))
    confidences.append(properties.get("ConfidenceID_2"))
    confidences.append(properties.get("ConfidenceID_3"))
    confidences.append(properties.get("ConfidenceID_4"))
    confidences.append(properties.get("ConfidenceID_5"))
    confidences.append(properties.get("ConfidenceID_6"))
    confidences.append(properties.get("ConfidenceID_7"))
    confidences.append(properties.get("ConfidenceID_8"))
    confidences.append(properties.get("ConfidenceID_9"))

    # Damage Status
    damageStatus.append(properties.get("Damage_Status"))
    damageStatus.append(properties.get("Damage_Status_2"))
    damageStatus.append(properties.get("Damage_Status_3"))
    damageStatus.append(properties.get("Damage_Status_4"))
    damageStatus.append(properties.get("Damage_Status_5"))
    damageStatus.append(properties.get("Damage_Status_6"))
    damageStatus.append(properties.get("Damage_Status_7"))
    damageStatus.append(properties.get("Damage_Status_8"))
    damageStatus.append(properties.get("Damage_Status_9"))

    # Main Damage Site Classes
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_2"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_3"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_4"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_5"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_6"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_7"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_8"))
    mainDamageSiteClasses.append(properties.get("Main_Damage_Site_Class_9"))

    newFeature["properties"] = {
        "SensorID": sensors,
        "SensorDate": sensorDates,
        "Confidence": confidences,
        "DamageStatus": damageStatus,
        "Main_Damage_Site_Class": mainDamageSiteClasses
    }
    newFeature["geometry"] = feature["geometry"]
    newFeature["type"] = "Feature",

    newFeatures.append(newFeature)


out_file = open("damage_gaza.json", "w")
json.dump({
    "type": "FeatureCollection",
    "features": newFeatures
},out_file)

out_file.close()


print('done')  



```
